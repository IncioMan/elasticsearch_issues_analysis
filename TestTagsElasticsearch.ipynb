```python
repo = Repo("/Users/alexincerti/Code/elasticsearch")
```

```python
git = repo.git
```

```python
tags = list(repo.tags)
```

```python
[tag.name for tag in tags]
```

```python
runs = 4
```

```python
result_folder = "/Users/alexincerti/Code/test_results/"
#search.aggregations.bucket
```

```python
for tag in tags[-5:-4]:
    tag_name = tag.name
    print(f"Working on tag {tag_name}")
    #
    try:
        git.checkout("-f", tag_name)
    except:
        print(f"Creating a branch for tag {tag_name} and checking it out...")
        git.checkout(tag_name, b=tag_name)
    #
    for run in range(runs):
        print(f"Run {run}...")
        process = subprocess.run(["/Users/alexincerti/Code/elasticsearch/gradlew",
                                  "-p", "/Users/alexincerti/Code/elasticsearch",
                                  ":server:test", 
                                  #"--tests",
                                  #"org.apache.lucene.queries.DoubleRandomBinaryDocValuesRangeQueryTests",
                                  #"org.elasticsearch.search.aggregations.bucket.missing.*",
                                 f"-Dorg.gradle.java.home={version_jdk[tag_name]}",
                                 "--build-cache"],
                                capture_output=True)
        #
        dest = f'{result_folder}{tag_name}/{run}'
        src = '/Users/alexincerti/Code/elasticsearch/server/build/reports/tests/test'
        if process.returncode == 0:
            if os.path.exists(dest):
                shutil.rmtree(dest)
            #create the folder
            shutil.copytree(src, dest)
        else:
            print(process.stdout)
            print(process.stderr)
print("Finished")
```
