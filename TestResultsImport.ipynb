```python
from bs4 import BeautifulSoup
import pandas as pd
import os
from pathlib import Path
```

```python
reports_location = "/Users/alexincerti/Code/test_results/"
```

```python
p = Path(reports_location)
versions = [f.name for f in p.iterdir() if f.is_dir()]
```

```python
class_test = "org.apache.lucene.queries.DoubleRandomBinaryDocValuesRangeQueryTests"
```

```python
tests = []
for version in versions:
    p = Path(f"{reports_location}{version}")
    runs = [f.name for f in p.iterdir() if f.is_dir()]
    for run in runs:
        with open(f"{reports_location}{version}/{run}/classes/{class_test}.html") as fp:
            soup = BeautifulSoup(fp)
            #print(soup.prettify())

        test_table = soup.find_all('table')[2]
        data = test_table.find_all("tr")

        for row in range(1,len(data)):
            tests.append((version, run, class_test,
                          data[row].find_all("td")[0].string,
                        data[row].find_all("td")[1].string,
                        data[row].find_all("td")[2].string))
```

```python
pd.DataFrame(tests, columns=["Tag", "Run", "Class", "Test", "Time", "Status"])
```
